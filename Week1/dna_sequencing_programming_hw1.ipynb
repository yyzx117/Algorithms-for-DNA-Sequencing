```python
def readGenome(filename):
    genome = ''
    with open(filename, 'r') as f:
        for line in f:
            # ignore header line with genome information
            if not line[0] == '>':
                genome += line.rstrip()
    return genome
genome = readGenome("lambda_virus.fa.1")
print(genome)
```

```python
#Q1: How many times does AGGT or its reverse complement (ACCT) occur in the lambda virus genome?
#Q2: How many times does TTAA or its reverse complement occur in the lambda virus genome?
def naive(p, t):
    occurrences = 0
    for i in range(len(t) - len(p) + 1):  # loop over alignments
        match = True
        for j in range(len(p)):  # loop over characters
            if t[i+j] != p[j]:  # compare characters
                match = False
                break
        if match:
            occurrences += 1  # all chars matched; record
    return occurrences

print(naive('AGGT', genome) + naive('ACCT', genome))
print(naive('TTAA', genome))
```

```python
#Q3: What is the offset of the leftmost occurrence of ACTAAGT or its reverse complement in the Lambda virus genome?
#Q4: What is the offset of the leftmost occurrence of AGTCGA or its reverse complement in the Lambda virus genome?
def naive(p, t):
    for i in range(len(t) - len(p) + 1):  # loop over alignments
        match = True
        for j in range(len(p)):  # loop over characters
            if t[i+j] != p[j]:  # compare characters
                match = False
                break
        if match:
            return i

print(min(naive('ACTAAGT', genome), naive('ACTTAGT', genome)))
print(min(naive('AGTCGA', genome), naive('TCGACT', genome)))
```

```python
#Q5: How many times does TTCAAGCC occur in the Lambda virus genome when allowing up to 2 mismatches?
def naive_2mm(p, t):
    occurrences = 0
    for i in range(len(t) - len(p) + 1):  # loop over alignments
        miss = 0
        match = True
        for j in range(len(p)):  # loop over characters
            if t[i+j] != p[j]:
              miss += 1
              if miss > 2:
                match = False
                break
        if match:
          occurrences += 1
    return occurrences

print(naive_2mm('ACTTTA', 'ACTTACTTGATAAAGT'))
print(naive_2mm('TTCAAGCC', genome))
```

```python
def readFastq(filename):
    sequences = []
    qualities = []
    with open(filename) as fh:
        while True:
            fh.readline()  # skip name line
            seq = fh.readline().rstrip()  # read base sequence
            fh.readline()  # skip placeholder line
            qual = fh.readline().rstrip() # base quality line
            if len(seq) == 0:
                break
            sequences.append(seq)
            qualities.append(qual)
    return sequences, qualities

sequences, qualities = readFastq('ERR037900_1.first1000.fastq')
print(qualities)
```

```python
#Q7: This dataset has something wrong with it; one of the sequencing cycles is poor quality.
#Report which sequencing cycle has the problem.  Remember that a sequencing cycle corresponds to a particular offset in all the reads.
#For example, if the leftmost read position seems to have a problem consistently across reads, report 0. If the fourth position from the left has the problem, report 3.
#Do whatever analysis you think is needed to identify the bad cycle.
import numpy as np

def phred33ToQ(quality):
  return ord(quality) - 33

def findByPos(qualities):
  qua = []
  for quality in qualities:
    num = 0
    total = 0
    for char in quality:
      num += 1
      q = phred33ToQ(char)
      total += q
    qua.append(float(total) / num)
  cycle = np.argmin(qua)
  return cycle
print(findByPos(qualities))
```
